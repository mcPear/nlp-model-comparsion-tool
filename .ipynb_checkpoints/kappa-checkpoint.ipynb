```python
import numpy as np
import sklearn.metrics as sm
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from preprocess_data import label_ordered_models, get_label_counts, confusion_matrix_preprocessed
```

```python
def get_cohen_data():
    models_data = confusion_matrix_preprocessed()
    model_preds_3=[]
    model_preds_4=[]
    model_names_3=[]
    model_names_4=[]
    for data in models_data:
        pred=data[0].iloc[:,1]
        name=data[1]
        label_count=data[2]
        if label_count==4:
            model_preds_4.append(pred)
            model_names_4.append(name)
        else:
            model_preds_3.append(pred)
            model_names_3.append(name)
    return model_preds_3, model_preds_4, model_names_3, model_names_4
```

```python
def get_kappa_matrix(labels, users):
    arr = np.array([
        [sm.cohen_kappa_score(labels[u1], labels[u2]) for u2 in range(4)]
        for u1 in range(4)
    ])
    df = pd.DataFrame(arr, columns=users, index=users)
    ax = sns.heatmap(df, vmin=0, vmax=1, annot=True)
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")
    # fix for mpl bug that cuts off top/bottom of seaborn viz
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show() # ta-da!
    
model_preds_3, model_preds_4, model_names_3, model_names_4 = get_cohen_data()
```

```python
get_kappa_matrix(model_preds_4, model_names_4)
```

```python
get_kappa_matrix(model_preds_3, model_names_3)
```
