```python
from preprocess_data import models, merge_all
import pandas as pd
import numpy as np
pd.options.display.max_colwidth = 10000
models = models()
```

```python
def unanimous_wrong(row):
    preds=[]
    for model in models:
        preds.append(row[model])
    preds=np.array(preds)
    return row.loc['Label'] != preds[0] and (preds == preds[0]).all()

def unanimous_good(row):
    preds=[]
    for model in models:
        preds.append(row[model])
    preds=np.array(preds)
    return row.loc['Label'] == preds[0] and (preds == preds[0]).all()

def nonunanimous(row):
    preds=[]
    for model in models:
        preds.append(row[model])
    set_=set(preds)
    return len(set_)>=2

def filter_true(column, df):
    return df[df[column]==True][['Tweet','Label']+models]

def add_interesting_cols(df):
    df['unanimous_wrong'] = df.apply(lambda x: unanimous_wrong(x), axis=1)
    df['unanimous_good'] = df.apply(lambda x: unanimous_good(x), axis=1)
    df['nonunanimous'] = df.apply(lambda x: nonunanimous(x), axis=1)
    return df

df_l4, df_l3 = merge_all() 

df_l4 = add_interesting_cols(df_l4)
df_l3 = add_interesting_cols(df_l3)

unanimous_wrong_l4 = filter_true('unanimous_wrong', df_l4)
unanimous_good_l4 = filter_true('unanimous_good', df_l4)
nonunanimous_l4 = filter_true('nonunanimous', df_l4)

unanimous_wrong_l3 = filter_true('unanimous_wrong', df_l3)
unanimous_good_l3 = filter_true('unanimous_good', df_l3)
nonunanimous_l3 = filter_true('nonunanimous', df_l3)
```
